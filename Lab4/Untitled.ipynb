<div><span style="font-size:14pt; font-weight:bold">Задание 1: </span> <span style="font-size:14pt">  Сравнение баллов IQ для различных тестовых групп с помощью блок-схемы </span></div>


a. Импортируем необходимые модулия:

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

b. Считываем данные:

```python
mydata = pd.read_csv('scores.csv')
mydata.head(7)
```

c. Преобразуем данные из столбцов в список с помощью метода tolist():

```python
group_a = mydata[mydata.columns[0]].tolist() 
group_b = mydata[mydata.columns[1]].tolist() 
group_c = mydata[mydata.columns[2]].tolist() 
group_d = mydata[mydata.columns[3]].tolist()
```

d. Выведиv переменные каждой группы на печать:

```python
print(group_a)
```

```python
print(group_b)
```

```python
print(group_c)
```

```python
print(group_d)
```

e. построение DataFrame:

```python
data = pd.DataFrame({'Groups': ['Group A'] * len(group_a) + ['Group B'] * 
len(group_b) + ['Group C'] * len(group_c) + ['Group D'] * len(group_d), 'IQ score': group_a + group_b + group_c + group_d})
data
```

f. Построение графика:

```python
plt.figure(dpi=150)

sns.set_style('whitegrid')

sns.boxplot('Groups', 'IQ score', data=data) 

sns.despine(left=True, right=True, top=True)
plt.title('IQ scores for different test groups')
plt.show() 
```

Из полученного бокс-графика можно сделать вывод, что IQ группы A больше, чем у других групп.

a. Считывание данных

```python
mydata = pd.read_csv("anage_data.csv")
mydata
```

b. Удаление лишнего столбца

```python
mydata = mydata.drop(columns = 'Unnamed: 0')
mydata
```

```python
longevity = 'Maximum longevity (yrs)'
mass = 'Body mass (g)'
data = mydata[mydata['Class'] == 'Mammalia']
data = data[np.isfinite(data[longevity]) & np.isfinite(data[mass]) & (data[mass] < 200000)] #isfinite - чек на NaN
data
```

 <div><span style="font-size:14pt"> Вывод </span></div>

```python
sns.set()
plt.figure(figsize=(10, 6), dpi=300)

sns.regplot(mass, longevity, data=data) 
plt.show()
```

 <div><span style="font-size:14pt"> Вывод: </span> существует линейная зависимость между массой тела и максимальной продолжительностью жизни для класса "Маммалия".</div>

Дополнительное задание: построить столбчатую диаграмму по типу комнат каждого из районов

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df_nyc = pd.read_csv('AB_NYC_2019.csv', index_col=0)
df_nyc.head()
```

```python
df_rooms = df_nyc[['neighbourhood_group', 'room_type', 'price']]
df_rooms['mean_price'] = df_rooms.groupby(['neighbourhood_group', 'room_type'])['price'].transform('mean')
df_rooms.rename(index=df_rooms["neighbourhood_group"]).drop(["neighbourhood_group", "price"], axis = 1)
```

```python
df = df_rooms.pivot_table(index = 'neighbourhood_group',
       columns = 'room_type',
       values = 'mean_price')
df
```

```python
import seaborn as sns
plt.style.use('seaborn')
sns.set_style("white")
sns.set_theme(style = "ticks")
```

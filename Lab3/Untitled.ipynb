```python
import numpy as np
import matplotlib.pyplot as plt
```

b. Создаем массивы для обозначения области определения двух 
надставленных областеɣ и соответствующие им области значениɣ:

```python
x = [0, 1, 2, 3, 4, 5]
y1 = [2, 6, 4, 8, 2, 4]
y2 = [2, 4, 2, -1, 4, -2]
```

c. Строим график

<div><span style="font-size:14pt; font-weight:bold">Задание 6: </span> <span style="font-size:14pt">Stacked Area Chart  </span></div>

<div><span style="font-size:12pt"> Нарисуйте 3 уровня лестницы с помощью Stacked Area Chart различными 
цветами, добавьте леɣблы и название графика </span></div>

```python
x = np.arange(0, 11, 1) 
y1 = np.array([i for i in x])
y2 = np.array([2*i for i in x])
y3 = np.array([4*i for i in x])

labels = ["y1", "y2", "y3"]
```

```python
plt.stackplot(x, y1, y2, y3, labels = labels)
plt.legend(loc = 'upper left')
plt.title('График')

plt.show()
```

```python
x = np.arange(0, 11, 1)
y_1, y_2, y_3 = [], [], []

for i in x:
    x[i]
    y_1.append(len(x))
    if i < len(x)/3:
        y_2.append(0)
        y_3.append(0)
    if i > len(x)/3 and i < len(x)/2:
        y_2.append(len(x))
        y_3.append(0)
    if i > len(x)/2:
        y_2.append(len(x))
        y_3.append(len(x))
```

```python
plt.stackplot(x, y_1, y_2, y_3, labels = labels)
plt.legend(loc = 'upper left')
plt.title('Конечный график')

plt.show()
```

Задание: сделать stacked area chart среднй цены по типам комнат по 5 районам

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df_nyc = pd.read_csv('AB_NYC_2019.csv', index_col=0)
df_nyc.head()
```

```python
df_rooms = df_nyc[['neighbourhood_group', 'room_type', 'price']]
df_rooms['mean_price'] = df_rooms.groupby(['neighbourhood_group', 'room_type'])['price'].transform('mean')
df_rooms.rename(index=df_rooms["neighbourhood_group"]).drop(["neighbourhood_group", "price"], axis = 1)
```

```python
df = df_rooms.pivot_table(index = 'neighbourhood_group',
       columns = 'room_type',
       values = 'mean_price')
df
```

```python
import seaborn as sns
plt.style.use('seaborn')
sns.set_style("white")
sns.set_theme(style = "ticks")
```

```python
df.plot.area(figsize = (20,9))
plt.title("Средняя цена по типам комнат по 5 районам", fontsize = 22, loc = 'left')
plt.xlabel("Районы", fontsize = 22)
plt.ylabel("Цена", fontsize = 22)
```

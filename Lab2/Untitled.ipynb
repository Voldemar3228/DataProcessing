<div><span style="font-size:12pt">При помощи Pie Charts создайте график, который отражает
континенты планеты и как подгруппы их страны и города. </span></div>

a. Добавляем интересующие нас библиотеки для реализации
данного задания:


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

b. Создаем массивы для дальнейшей работы программы

```python
source_data={'continents':['Africa',
                           'Antarctida',
                           'Europe&Asia',
                           'Australia',
                           'North America',
                           'South America'],
'countries_count':[68, 3, 99, 6, 23, 15],
'cities_count':[68, 15, 136, 86, 100, 89]}

# Создания Data Frame
df=pd.DataFrame(source_data,columns=['continents',
                                     'countries_count',
                                     'cities_count'])
```

c. Демонстрация полученного Data Frame

```python
df
```

d. Отрисовка полученного графика

```python
# Настройка цветовой гаммы
cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(5000)*2)
inner_colors = outer_colors


# Настройка масштаба графика
plt.figure(figsize=(8,10))

# Отрисовка внешней части pie
plt.pie(df.countries_count,
        labels = df.iloc[0:6, 0], 
        startangle=90,
        pctdistance =0.8,
        colors=outer_colors,
        autopct = '%1.1f%%',
        radius= 0.9,
        labeldistance=1.05,
        textprops ={'fontweight': 'bold','fontsize':13},
        wedgeprops = {'linewidth' : 3, 'edgecolor' : "w" } )

# Отрисовка внутренней части pie - убраны названия, чтобы не засорять график
plt.pie(df.cities_count,
        startangle=90,
        pctdistance =0.7,
        colors=inner_colors,
        autopct = '%1.1f%%',
        radius= 0.5,
        textprops ={'fontweight': 'bold' ,'fontsize':13},
        wedgeprops = {'linewidth' : 3, 'edgecolor' : "w" } )

# Отрисовка белого выреза внутри
centre_circle = plt.Circle((0,0), 0.2, fc='white')
fig= plt.gcf()
fig.gca().add_artist(centre_circle) # adding the centre circle
```
